```python
company=soup.find_all("div",class_="company-content-wrapper")
```

```python
for i in company:
  print(i.find('a',class_="review-count").text.strip())
```

```python
name=[]
rating=[]
reviews=[]

for i in company:
  name.append(i.find('h2').text.strip())
  rating.append(i.find('p',class_="rating").text.strip())
  reviews.append(i.find('a',class_="review-count sbold-Labels").text.strip())
```

```python
reviews
```

```python
ctype=[]
location=[]
old=[]
employees=[]

for i in company:
  ctype.append((i.find_all('p',class_="infoEntity sbold-list-header")[0].text.strip()))
  location.append((i.find_all('p',class_="infoEntity sbold-list-header")[1].text.strip()))
  old.append((i.find_all('p',class_="infoEntity sbold-list-header")[2].text.strip()))
  employees.append((i.find_all('p',class_="infoEntity sbold-list-header")[3].text.strip()))
```

```python
employees
```

```python
d={'name': name,
   'rating':rating,
   'reviews':reviews,
   'ctype':ctype,
   'location':location,
   'old':old,
   'employees':employees
}

df=pd.DataFrame(d)
df
```

```python
final=pd.DataFrame()
for j in range(1,334):
  url="https://www.ambitionbox.com/list-of-companies?pages=1&page={}".format(j)
  headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
  webpage=requests.get(url,headers=headers).text
  soup=BeautifulSoup(webpage,"html")
  company=soup.find_all("div",class_="company-content-wrapper")

  name=[]
  rating=[]
  reviews=[]
  ctype=[]
  location=[]
  old=[]
  employees=[]


  for i in company:
    try:
      name.append(i.find('h2').text.strip())
    except:
      name.append(np.nan)
    try:
      rating.append(i.find('p',class_="rating").text.strip())
    except:
      rating.append(np.nan)
    try:
    
      reviews.append(i.find('a',class_="review-count sbold-Labels").text.strip())
    except:
      reviews.append(np.nan)
    try:  
      ctype.append((i.find_all('p',class_="infoEntity sbold-list-header")[0].text.strip()))
    except:
      ctype.append(np.nan)
    try:
      location.append((i.find_all('p',class_="infoEntity sbold-list-header")[1].text.strip()))
    except:
      location.append(np.nan)
    try:  
      old.append((i.find_all('p',class_="infoEntity sbold-list-header")[2].text.strip()))
    except:
      old.append(np.nan)
    try:
      employees.append((i.find_all('p',class_="infoEntity sbold-list-header")[3].text.strip()))
    except:
      employees.append(np.nan)
  d={'name': name,
   'rating':rating,
   'reviews':reviews,
   'ctype':ctype,
   'location':location,
   'old':old,
   'employees':employees
  }

  df=pd.DataFrame(d)

  final=pd.concat([final,df],ignore_index=True)
```
